Given an array, we have to find the smallest element in the array.

```python
def sortArr(arr):
    arr.sort()
    
    return arr[0]

if __name__ == "__main__":
    arr1 = [2,3,1,4,6,8,0,5,7,9]
    print("The smallest number in the array is: ", sortArr(arr1))
```

Given an array, we have to find the largest element in the array.

```python
def LargerArr(arr):
    arr.sort()
    
    return arr[-1]

if __name__ == "__main__":
    arr1 = [2,3,1,4,6,8,0,5,7,9]
    print("The largest number in the array is: ", LargerArr(arr1))
```

Given an array, find the second smallest and second largest element in the array. Print ‘-1’ in the event that either of them doesn’t exist.

```python
def ssl(arr):
    arr.sort()
    
    return arr[1], arr[-2]

if __name__ == "__main__":
    arr1 = [2,3,1,4,6,8,0,5,7,9]
    print("The second smallest and second largest number in the array is: ", ssl(arr1))
```

You are given an array. The task is to reverse the array and print it.

```python
class Solution:
    def reverseArray(self, arr):
        n = len(arr)
        ans = [0]*n
        for i in range(n):
            ans[i] = arr[n-1-i]
        return ans

if __name__ == "__main__":
    obj = Solution()
    arr = [1, 2, 3, 4, 5]
    result = obj.reverseArray(arr)
    print("Reversed Array:", result)
    
```
